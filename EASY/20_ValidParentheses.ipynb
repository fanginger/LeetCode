### 20. Valid Parentheses
Given a string s containing just the characters '(', ')', '{', '}', '[' and ']', determine if the input string is valid.

### Stack (LIFO)

- 先進後出！
- 只能從堆疊的`最頂端`存取資料
- 只能從堆疊的`最頂端`新增或刪除資料
- 資料的存取必須符合`後進先出`(Last In First Out, LIFO)

這題核心重點，將左邊的都先加進去，則如果出現為右邊則必須向左邊匹配上
所以符合stack的概念，後進的東西需要先出去先匹配上

```python
class Solution:
    def isValid(self, s: str) -> bool:
        OP , REV, stack = {'{',"[",'('} , {'}':'{',')':'(',']':'['} , []
				
        for c in s:
            if c in OP:
                stack.append(c)
            else:
                if len(stack) == 0:
                    return False
                if REV[c] == stack[-1]:
                    stack.pop()
                else:
                    return False
        return len(stack) == 0
```
