### 658. Find K Closest Elements
Given a sorted integer array arr, two integers k and x, return the k closest integers to x in the array. The result should also be sorted in ascending order.

An integer a is closer to x than an integer b if:

- |a - x| < |b - x|, or
- |a - x| == |b - x| and a < b


想法：題目要找離x最近的k個元素，另外這list是有經過排列。那就可以從兩頭開始去篩除，因為距離x最遠的數字一定是在首尾。所以問題就比較簡單。持續比較兩個首尾，直到長度剩下k

```python
class Solution:
    def findClosestElements(self, arr: List[int], k: int, x: int) -> List[int]:
        temp = arr
        if len(temp) == k:
            return temp
        else:
            for i in range(len(arr)-1):
                while len(temp) > k:
                    if abs(temp[i] - x) < abs(temp[-i-1] - x):
                        temp.pop(-i-1)
                    elif abs(temp[i] - x) == abs(temp[-i-1] - x) and temp[i]<temp[-i-1]:

                        temp.pop(-i-1)
                    else:
                        temp.pop(i)

                return temp
```
